```python
from keras.models import Sequential, Model
from keras.preprocessing.image import ImageDataGenerator
from keras import layers
import numpy as np
from time import time
```

```python
from keras.preprocessing import image
import os 
import matplotlib.pyplot as plt

# View image example
data_dir = '/Users/Manuel/Desktop/Projects/RPS/my_data'
dst_dir = '/Users/Manuel/Desktop/Projects/RPS/type_my_data'
labels=['paper', 'rock', 'scissor']
target_size=(128,128)

images = []
f, axarr = plt.subplots(1,3, figsize=(10, 2))
for i in range(len(labels)):
    aux_dir = os.path.join(data_dir, labels[i])
    img = image.load_img(os.path.join(aux_dir, os.listdir(aux_dir)[10]), target_size=target_size)
    images.append(img)
    axarr[i].imshow(img)
    axarr[i].title.set_text(labels[i])

img_tensor = image.img_to_array(img)
print(img_tensor.shape)
```

```python
resize_data(data_dir, labels, target_size=(128,128))
#add_pepper_noise_data(data_dir, labels, size=(128,128,3))
create_folders(data_dir, dst_dir, labels, done=True)
```

```python
height =  target_size[0]
width = target_size[1]
channels = 1 # Due to color_mode = "grayscale"
num_classes = 3
```

```python
batch_size = 16

train_datagen = ImageDataGenerator(rescale=1./255)
train_gen = train_datagen.flow_from_directory(
    os.path.join(dst_dir,'train'),
    target_size=target_size,
    batch_size=batch_size,
    class_mode='categorical',
    color_mode = "grayscale",
    classes=labels
)

val_datagen = ImageDataGenerator(rescale=1./255)
val_gen = val_datagen.flow_from_directory(
    os.path.join(dst_dir,'validation'),
    target_size=target_size,
    batch_size=batch_size,
    class_mode='categorical',
    color_mode = "grayscale",
    classes=labels
)

test_datagen = ImageDataGenerator(rescale=1./255)
test_gen = val_datagen.flow_from_directory(
    os.path.join(dst_dir,'validation'),
    target_size=target_size,
    batch_size=batch_size,
    class_mode='categorical',
    color_mode = "grayscale",
    classes=labels
)
```

```python
# Model definition 
model = Sequential()

model.add(layers.Conv2D(16,(3,3),activation='relu',input_shape=(height, width, channels)))
model.add(layers.Conv2D(32,(3,3),activation='relu'))
model.add(layers.MaxPooling2D(2))

model.add(layers.Conv2D(64,(3,3),activation='relu'))
model.add(layers.Conv2D(128,(3,3),activation='relu'))
model.add(layers.MaxPooling2D(2))

model.add(layers.Conv2D(256,(3,3),activation='relu'))
model.add(layers.Conv2D(128,(3,3),activation='relu'))
model.add(layers.MaxPooling2D(2))

#model.add(layers.BatchNormalization())

model.add(layers.Conv2D(64,(3,3),activation='relu'))
model.add(layers.Conv2D(32,(3,3),activation='relu'))
model.add(layers.GlobalAveragePooling2D())

#model.add(layers.BatchNormalization())

model.add(layers.Dense(64,activation='relu'))
model.add(layers.Dense(num_classes, activation='softmax'))

model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['acc'])

model.summary()
```

```python
epochs = 12
start = time()
history = model.fit(train_gen, epochs=epochs, validation_data=val_gen, verbose=0)
end = time()
print('Seconds per epoch: ' + str((end-start)/epochs) + ' // Epochs: ' + str(epochs) + ' // Seconds: ' + str(end-start))
```

```python
history_dict = history.history

val_acc = history_dict['val_acc']
acc = history_dict['acc']
val_loss = history_dict['val_loss']
loss = history_dict['loss'] 

epochs_plt = range(1,len(val_acc) + 1)

plt.plot(epochs_plt,val_acc,'b',label='Validation Acc')
plt.plot(epochs_plt,acc,'ro',label='Acc')
plt.title('Accuracy')
plt.xlabel('Epochs')
plt.legend()
plt.show()

plt.plot(epochs_plt,val_loss,'b',label='Validation Loss')
plt.plot(epochs_plt,loss,'ro',label='Loss')
plt.title('Loss')
plt.xlabel('Epochs')
plt.legend()
plt.show()
```

```python
test_loss, test_acc = model.evaluate(test_gen)
```

```python
import shutil
import os

# Terrible function, will not copy the last couple of images
def create_folders(data_dir, dst_dir, labels, done=False, sets={'train':0.6,'validation':0.2,'test':0.2}):
    
    if (done != True):
        if not os.path.exists(dst_dir):
            os.mkdir(dst_dir)    

        for set_type in sets.keys():
            aux_dir = os.path.join(dst_dir, set_type)
            if not os.path.exists(aux_dir):
                os.mkdir(aux_dir)

        for set_type in sets.keys():
            for label in os.listdir(data_dir):
                aux_dir = os.path.join(os.path.join(dst_dir, set_type),label)
                if not os.path.exists(aux_dir):
                    os.mkdir(aux_dir)

        N_sizes = {'train':int(0), 'test':int(0), 'validation':int(0)}
        for label in os.listdir(data_dir):
            # Working in data -> label folder (1)
            data_aux_dir = os.path.join(data_dir, label)

            # Define the images and number them in N
            image_list = os.listdir(data_aux_dir)
            N = len(image_list)

            # Defining limits for each set in number of images in label
            sets_lim = {}
            aux_inf = 0
            aux_sup = 0
            for set_type,set_prob in sets.items():
                aux_sup = aux_inf + np.floor(N * set_prob) 
                sets_lim[set_type] = [aux_inf, aux_sup]
                N_sizes[set_type] = N_sizes[set_type] +  aux_sup - aux_inf
                aux_inf = aux_sup


            # Take each image randomly
            for k in np.random.permutation(range(N)):
                # Iterate over all sets types to find the correct one
                for set_type, set_lim in sets_lim.items():
                    # Check in what set it goes
                    if (k >= set_lim[0]) and (k < set_lim[1]):

                        # Copy data
                        aux_dir_data = os.path.join(os.path.join(data_dir, label), image_list[k])
                        aux_dir_dest = os.path.join(os.path.join(os.path.join(dst_dir, set_type), label), image_list[k])
                        shutil.copyfile(aux_dir_data,aux_dir_dest) 

        # Change values into int because of floor function
        for key,val in N_sizes.items():
            N_sizes[key] = int(val)
    else:
        print('Data has already been copied. Not copying again')
```

```python
from PIL import Image
# Resize all data 
def resize_data(data_dir, labels, target_size):
    for label in labels:
        aux_dir = os.path.join(data_dir, label)
        for image_name in os.listdir(aux_dir):
            aux_dir2 = os.path.join(aux_dir, image_name)
            img = Image.open(aux_dir2)
            img = img.resize(target_size, Image.ANTIALIAS)
            img.save(aux_dir2)

def add_pepper_noise_data(data_dir, labels, size, noise_param=0.07):
    for label in labels:
        aux_dir = os.path.join(data_dir, label)
        for image_name in os.listdir(aux_dir):
            
            image_dir = os.path.join(aux_dir, image_name)
            img = Image.open(image_dir)
            img_tensor = image.img_to_array(img)
            img_tensor = np.mod(img_tensor + np.matmul(((np.random.rand(128,128,1) < noise_param)* 255), np.ones((1,3))) ,256)
            image.save_img(image_dir, img_tensor)
```
